#Paquetes necesarios

```python
import cv2  
import numpy as np
import matplotlib.pyplot as plt
```

Carga imagen y convierte a RGB

```python
img = cv2.imread('Monedas.jpg') 
print(img.shape)
#OpenCV lee las imgenes en BGR, por lo que convertimos para visualizr RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.axis("off")
plt.imshow(img_rgb) 
plt.show()
```

Convierte a gris y muestra histograma

```python
#Convierte a gris y muestra
img_gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#Clculo del histograma de una imagen en escala de grises
hist = cv2.calcHist([img_gris], [0], None, [256], [0, 256])

plt.figure()
plt.axis("off")
plt.imshow(img_gris, cmap='gray')

# Histograma sin normalizar
plt.figure()
plt.subplot(1, 2, 1)
plt.title("Histograma")
plt.xlabel("Bins")
plt.ylabel("# pxeles")
plt.plot(hist)
plt.xlim([0, 256])
```

Cuenta elementos tras umbralizar

```python
#Dos umbralizados. Te invito a probar con parmetros, aplicar filtro previo, etc.
umbral = 220
#Umbralizacin binaria invertida, objetos en blanco
res,img_th1 = cv2.threshold(img_gris,umbral,255,cv2.THRESH_BINARY_INV)
# Otsu
ret2,img_th2 = cv2.threshold(img_gris,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

plt.subplot(1, 3, 1)
plt.axis("off")
plt.imshow(img_gris,cmap='gray') 
plt.title('Original')

plt.subplot(1, 3, 2)
plt.axis("off")
plt.imshow(img_th1,cmap='gray') 
plt.title('FIJO invertida')

plt.subplot(1, 3, 3)
plt.axis("off")
plt.imshow(img_th2,cmap='gray') 
plt.title('OTSU invertida')

```

Bsqueda de componentes y sus contornos

```python
#Localiza contornos en imagen obtenida con umbral fijo
#findContours est diseada para imgenes con la figura en blanco y el fondo negro
#La imagen de entrada debe ser de un canal y 8 bits eccepto en los modos RETR_CCOMP o RETR_FLOODFILL
#hierarchy contiene informacin sobre el nivel del contorno, relaciones paterno-filiales

#Todos los contornos, externos e internos
contornos, hiptaerarchy = cv2.findContours(
    img_th1, #imagen
    cv2.RETR_TREE, #Modo de recuperacin (lista, rbol, nivel superior)
    cv2.CHAIN_APPROX_SIMPLE #Mtodo de proximacin del contorno
    )
#Slo contornos externos
contornos2, hierarchy2 = cv2.findContours(img_th1, cv2.RETR_EXTERNAL , cv2.CHAIN_APPROX_SIMPLE)

#Cada vez que quiere pintar convierte img para no tener restos
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Dibuja sobre la imagen de entrada todos los contornos en verde
cv2.drawContours(img_rgb, contornos, -1, (0,255,0), 3)
plt.subplot(131)
plt.axis("off")
plt.imshow(img_rgb) 
plt.title('Todos los contornos')

# Dibuja sobre la imagen de entrada slo contornos externos
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.drawContours(img_rgb, contornos2, -1, (0,0,255), 3)
plt.subplot(132)
plt.axis("off")
plt.imshow(img_rgb) 
plt.title('Slo externos')

#Dibuja contornos externos en imagen vaca
#Muestra informacin
#Imagen vaca
img_cont = np.zeros(img_rgb.shape)
for c in contornos2:
    #rea
    area = cv2.contourArea(c)

    #Dimensin mnima (til en ocasiones)
    if area > 10:
        #Permetro
        perimetro = cv2.arcLength(c,True)
        #Contenedor alineado con ejes de la imagen
        x,y,w,h = cv2.boundingRect(c)
        #Mnimo contenedor ajustado para el contorno
        rect = cv2.minAreaRect(c)
        #Mnimo crculo que lo contiene
        (cx,cy),radio = cv2.minEnclosingCircle(c)
        #Elipse ajustada
        if c.shape[0] > 5:
            elipse = cv2.fitEllipse(c)
            #Para determinadas tareas nos interesar mostrar estos valores
            #print(area, perimetro, rect, cx,cy,radio, elipse)

        #Dibuja los contornos
        cv2.drawContours(img_cont, [c], -1, (255,255,255), -1)

plt.subplot(133)
plt.axis("off")
plt.imshow(img_cont) 
plt.title('Rellenos')
plt.show()
```
